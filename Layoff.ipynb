```python
df = pd.read_excel("Tech_Layoff_Cleaned.xlsx")
```

```python
df
```

```python
# drop last row
df = df[:-1]
df.shape
```

```python
# remove header rows at the start of each page in the pdf file

df = df[df['Company'] != 'Company']
df.shape
```

```python
df.info()
# need to convert some columns to int, float and datetime
```

```python
df = df.rename(columns={'# Laid Off': 'laid_off', "$ Raised (mm)": "money_raised"})
```

```python
# converting NaN in laid_off column to 0 first in order to change the type to int. 

df["laid_off"] = df.laid_off.fillna(0).astype("int")
```

```python
df["money_raised"] = df.money_raised.astype("float")
```

```python
df["%"] = df["%"].astype("float")
```

```python
df["%"] = df["%"]*100
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
```

```python
# sort values by number of employees laid off

df.sort_values("laid_off", ascending=False).head(20)

# Uber - 2 rounds of layoff a week aprt
# Bytedance - 2 rounds of layoff in 2021
```

```python
# sort values by number of employees laid off in 2020, 2021 and 2022 (to date)

df['year'] = df.Date.dt.year

df.groupby("year").laid_off.sum()

# huge number in 2020, slowed down in 2021 considerably before the layoffs increase sharply again in just the first 7-8 months or so this year
```
